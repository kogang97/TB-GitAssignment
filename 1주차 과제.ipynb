```python
# 패키지 import
import pandas as pd
import numpy as np
```

### 2) 상관관계 확인

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.heatmap(df.corr(),cmap='summer')
plt.show()
```

```python
df.corr()sns.pairplot(df)
```

```python
df.corr()
```

```python
df_corr=df.corr()
```

```python
#상관관계가 0.3보다 큰 것 찾기
df_corr[df_corr>0.3]
```

## 상관관계 설명
elementary_school_count와 kindergarten_count, university_count, nursing_home_count가 양의상관관계가 있다.
이를 통해, 교육기관끼리 모여있음을 알 수 있다. 
또한, 교육기관과 요양원의 상관관계 또한 높다. 이를 통해, 교육기관이 모여있는 곳에 요양원도 있음을 알 수 있다. 

```python
#상관관계가 0.3보다 큰 것 찾기
df_corr[df_corr<-0.3]
```

음의 상관관계를 보면, 교육기관과 노인인구비율은 음의 상관관계를 가지며, 반비례함을 알 수 있다.노인인구비율은 요양원과도 음의 상관관계를 가지는데, 이는 요양원이 비교적 비효율적인 장소에 위치함을 알 수 있다.

## 3) 유의미한 시각화

```python
sns.pairplot(df, vars=["elementary_school_count", "kindergarten_count"])
```

초등교육기관과 유치원이 모여있음을 알 수 있음

```python
plt.scatter(x=df['elementary_school_count'],y=df['university_count']) 
plt.show()
```

이상치를 제외하면, 점들이 모여있는 것을 알 수 있다.

```python
g = sns.pairplot(df,
                 x_vars=["elementary_school_count", "university_count"],
                 y_vars=["kindergarten_count", "academy_ratio"])
```

가장 상관관계가 높은것은 유치원과 초등학교의 수이다. 왼쪽 상단을 보면 알 수 있다.

```python
df.iloc[0:,[12,13]]
```

```python
sns.lmplot(x="elderly_population_ratio", y="elderly_alone_ratio", data=df.iloc[0:,[12,13]])
```

노인 인구비율과 독거노인 비율의 상관관계가 굉장히 높음을 알 수 있다. 더 나아가, 독거노인 비율 / 노인 인구비율이라는 새로운 파생변수를 만들 수 있을 것 같다.

# group 별로 정리하기

```python
df2=df.groupby(['province_x']).sum()
```

```python
df2
```

```python
plt.bar(df2.index,df2['confirmed'])
plt.xticks(rotation=45)
```

대구에서 확진자수가 많음을 알 수 있다.

```python
sns.lmplot(x="elderly_population_ratio", y="confirmed", data=df2.iloc[0:,[6,1]])
```

결측치를 제외하면, 노인비율이 많다면 확진자수가 많음을 알 수 있다. 이를 통해 
노인비율/확진자수 이라는 파생새로운 파생변수를 만들 수 있을 것 같다.
